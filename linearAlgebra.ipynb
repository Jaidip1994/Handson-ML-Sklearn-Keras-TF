```python
# Plotting Vectors 
u = np.array([2,5])
v = np.array([3,1])

plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
# plotting Scatter Plot
x_cor, y_cor = zip(u,v)
plt.title('Vector as a Scatter Plot')
plt.scatter(x_cor, y_cor, color = ["r", "b"])
plt.axis([0,9,0,6])
plt.grid()

# depictation of Vectors as as a Arrow in 2D
def plot_vector_2d(vector2d, origin = [0,0], **options):
    plt.arrow(origin[0], origin[1], vector2d[0], vector2d[1], head_width = 0.2, length_includes_head = True, **options)
plt.subplot(1,2,2)
plt.title('Depictation of Vectors as as a Arrow in 2D')
plot_vector_2d(u, color = "r")
plot_vector_2d(v, color = "b")
plt.axis([0,9,0,6])
plt.grid()
plt.show()

# plottling vector in 3D Space

a = np.array([1, 2, 8])
b = np.array([5, 6, 3])

def plot_vectors3d(ax, vectors3d, z0, **options):
    for v in vectors3d:
        x, y, z = v
        ax.plot([x,x], [y,y], [z0, z], color="gray", linestyle='dotted', marker=".")
    x_coords, y_coords, z_coords = zip(*vectors3d)
    ax.scatter(x_coords, y_coords, z_coords, **options)

subplot3d = plt.subplot(111, projection='3d')
subplot3d.set_zlim([0, 9])
plot_vectors3d(subplot3d, [a,b], 0, color=("r","b"))
plt.show()
```
